#Calculadora con las funciones basicas hecha con OPP kafla_whats.jpg

```python
#Definir al objeto "Calculadora" por medio de clases
class Calculadora:
  def __init__(self, Num_1, Num_2):
    self.Num_1 = float(Num_1)
    self.Num_2 = float(Num_2)
#Escribimos cada instancia dependiendo de la operacion que queremos hacer
  def sumar(self):
    print(f'El resultado de sumar {self.Num_1} mas {self.Num_2} es igual a {self.Num_1 + self.Num_2}')

  def resta(self):
    print(f'El resultado de restar {self.Num_1} menos {self.Num_2} es igual a {self.Num_1 - self.Num_2}')

  def multiplicacion(self):
    print(f'El resultado de multiplicar {self.Num_1} por {self.Num_2} es igual a {self.Num_1 * self.Num_2}')

  def division(self):
     print(f'El resultado de dividir {self.Num_1} entre {self.Num_2} es igual a {self.Num_1 / self.Num_2}')


#Iniciar con un loop el programa preguntando al usuario si desea utilizar la calculadora
Loop = input('Desea utilizar la calculadora hecha con OOP? [Y/n]: ')
while Loop == 'Y' or Loop == 'y':
  #Definir los valores que se ingresaran en el "Objeto"
  Num_1 = float(input('Ingrese el primer numero: '))
  Num_2 = float(input('Ingrese el segundo numero: '))
  #Llamamos al "Objeto"
  Mycalculadora = Calculadora(Num_1, Num_2)
  #Creamos un menu con distintas opciones
  Menu = int(input('Que operacion desea realizar?: 1) Suma, 2) Resta, 3) Multiplicacion, 4) Division: '))
  #Dependiendo de la opcion llamada se hara una operacion distinta
  if Menu == 1:
    Mycalculadora.sumar()
  elif Menu == 2:
    Mycalculadora.resta()
  elif Menu == 3:
    Mycalculadora.multiplicacion()
  elif Menu == 4:
    Mycalculadora.division()

  #Preguntar si desea continuar o no el ciclo
  Loop = input('Desea continuar utilizando la calculadora? [Y/n]: ')




```
